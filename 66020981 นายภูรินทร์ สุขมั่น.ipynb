### Example 3.6 shows an example of SVM classification for breast cancer

```python
#Example 3.6 Python SVM Breast Cancer Classifications
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

cancer = load_breast_cancer()
X = cancer.data # All of the features
y = cancer.target # All of the labels

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =
0.2, random_state = 20)

clf = SVC()
clf.fit(X_train, y_train)

#Prediction
y_predict = clf.predict(X_test)

#Print Confusion Matrix and Classification Report
from sklearn.metrics import classification_report, confusion_matrix
cm = np.array(confusion_matrix(y_test, y_predict, labels=[1,0]))
confusion = pd.DataFrame(cm, index=['is_cancer', 'is_healthy'],
 columns=['predicted_cancer','predicted_healthy'])
print(confusion)
print(classification_report(y_test, y_predict))
```

### แบบฝึกหัด 3.4
ปรับเปลี่ยนโปรแกรม Python จากตัวอย่างที่ 3.6 ให้สามารถพล็อตฮิสโตแกรมของรัศมี, ขนาด, พื้นผิว, และความเรียบ
 (radius,size, texture, and smoothness)
ของจุดข้อมูลทั้งหมด

```python
# เขียนต้องนี้
# Example 3.6 Python SVM Breast Cancer Classifications - Modified for Histograms
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

cancer = load_breast_cancer()
X = cancer.data # All of the features
y = cancer.target # All of the labels

# เลือกฟีเจอร์ที่ต้องการ (radius, size, texture, and smoothness)
# คอลัมน์ที่เลือกคือ 0 (radius), 1 (size), 2 (texture), และ 3 (smoothness)
selected_features = X[:, [0, 1, 2, 3]]

# พล็อตฮิสโตแกรมของคุณลักษณะที่เลือก
import matplotlib.pyplot as plt

# กำหนดชื่อสำหรับคุณลักษณะที่เลือก
feature_names = ['radius', 'size', 'texture', 'smoothness']

# พล็อตฮิสโตแกรม
plt.figure(figsize=(10, 8))
for i, feature_name in enumerate(feature_names):
    plt.subplot(2, 2, i + 1)
    plt.hist(selected_features[:, i], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {feature_name}')
    plt.xlabel(feature_name)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# แบ่งข้อมูลเป็นชุดฝึกและทดสอบ
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)

clf = SVC()
clf.fit(X_train, y_train)

# Prediction
y_predict = clf.predict(X_test)

# Print Confusion Matrix and Classification Report
from sklearn.metrics import classification_report, confusion_matrix
cm = np.array(confusion_matrix(y_test, y_predict, labels=[1, 0]))
confusion = pd.DataFrame(cm, index=['is_cancer', 'is_healthy'],
                         columns=['predicted_cancer', 'predicted_healthy'])
print(confusion)
print(classification_report(y_test, y_predict))
```

### แบบฝึกหัด 3.5
ปรับเปลี่ยนโปรแกรม Python จากตัวอย่างที่ 3.7 ให้ทำการฝึกโมเดล Naive Bayes ก่อน บันทึกโมเดลลงในไฟล์, โหลดโมเดลจากไฟล์ และทำนายผลด้วยโมเดลนั้น

```python
# doing here
# Example 3.7 Naive Bayes Iris - Save and Load Model
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
import joblib

# โหลดข้อมูล Iris
X, y = load_iris(return_X_y=True)

# สร้างโมเดล Naive Bayes
clf = GaussianNB()

# ฝึกโมเดล
clf.fit(X, y)

# บันทึกโมเดลลงในไฟล์
joblib.dump(clf, 'naive_bayes_iris_model.pkl')

# โหลดโมเดลจากไฟล์
loaded_clf = joblib.load('naive_bayes_iris_model.pkl')

# ทำนายผลด้วยโมเดลที่โหลด
p = loaded_clf.predict([[5.0, 3.4, 1.5, 0.4]])

# แสดงผลการทำนาย
print(p)
```

### แบบฝึกหัด 3.7
ปรับเปลี่ยนโปรแกรม Python จากตัวอย่างที่ 3.10 ให้ทำการวิเคราะห์องค์ประกอบหลัก (PCA) กับข้อมูลมะเร็งเต้านม

```python
# coding
# Example: PCA on Breast Cancer Data
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA

# โหลดข้อมูลมะเร็งเต้านม
cancer = datasets.load_breast_cancer()
X = cancer.data  # คุณลักษณะทั้งหมด
y = cancer.target  # เลเบล (0 = healthy, 1 = cancer)

# พล็อตข้อมูลมะเร็งเต้านม
f = plt.figure(1)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Original Breast Cancer Data')
f.show()

# ทำการวิเคราะห์องค์ประกอบหลัก (PCA)
pca = PCA(n_components=3)
pca.fit(X)
X1 = pca.transform(X)

# พล็อตข้อมูลหลังจาก PCA
g = plt.figure(2)
plt.scatter(X1[:, 0], X1[:, 1], c=y)
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.title('PCA of Breast Cancer Data')
g.show()
```

```python
# coding
# Example: Decision Tree Classification on Wine Dataset
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# โหลดข้อมูลไวน์
X, y = load_wine(return_X_y=True)

# แบ่งข้อมูลเป็นชุดฝึกและทดสอบ
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)

# สร้างและฝึกโมเดลต้นไม้การตัดสินใจ
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

# ทำนายผลสำหรับชุดทดสอบ
y_pred = clf.predict(X_test)

# คำนวณจำนวนจุดที่ทำนายถูกต้อง
N = y_test.shape[0]
C = (y_test == y_pred).sum()

# แสดงผลลัพธ์
print("Total points: %d Correctly labeled points: %d" % (N, C))
```

```python
# coding
# Example: Random Forest on Diabetes Dataset
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# โหลดข้อมูลเบาหวาน
diabetes = load_diabetes()
X = diabetes.data  # คุณลักษณะทั้งหมด
y = diabetes.target  # เป้าหมาย (ค่าตัวเลขที่เกี่ยวข้องกับระดับน้ำตาลในเลือด)

# แบ่งข้อมูลเป็นชุดฝึกและทดสอบ
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)

# สร้างและฝึกโมเดล Random Forest
clf = RandomForestClassifier(random_state=0)
clf.fit(X_train, y_train)

# ทำนายผลสำหรับชุดทดสอบ
y_pred = clf.predict(X_test)

# คำนวณจำนวนจุดที่ทำนายถูกต้อง
N = y_test.shape[0]
C = (y_test == y_pred).sum()

# แสดงผลลัพธ์
print("Total points: %d Correctly labeled points: %d" % (N, C))
```

### แบบฝึกหัด 3.10
ปรับเปลี่ยนโปรแกรม Python จากตัวอย่างที่ 3.14 ให้ทำการจำแนกประเภทบนข้อมูลเบาหวานของ Scikit-Learn (https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_diabetes.html#sklearn.datasets.load_diabetes)

Citations:
[1] https://scikit-learn.org/stable/modules/

```python
# coding here!!!
# Example: Comparison of Different Classifiers on Diabetes Dataset
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import numpy as np

# ชื่อของโมเดล
names = [
    "SVM", "Naive Bayes", "LDA", "QDA", "Decision Tree", "Random Forest",
    "Nearest Neighbors", "Neural Networks"
]

# สร้างโมเดล
classifiers = [
    SVC(),
    GaussianNB(),
    LinearDiscriminantAnalysis(),
    QuadraticDiscriminantAnalysis(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    KNeighborsClassifier(),
    MLPClassifier(alpha=1, max_iter=1000)
]

# โหลดข้อมูลเบาหวาน
diabetes = load_diabetes()
X = diabetes.data  # คุณลักษณะทั้งหมด
y = diabetes.target  # เป้าหมาย (ระดับน้ำตาลในเลือด)

# เนื่องจากเป็นงานการทำนายค่าตัวเลข (Regression) เราจะแบ่งข้อมูลเป็น 2 คลาส
# แปลงข้อมูล y เป็นคลาสเพื่อใช้ในการจำแนกประเภท
y_class = np.digitize(y, bins=[100])  # แปลงค่าให้เป็น 2 คลาส (ต่ำ/สูง)

# แบ่งข้อมูลเป็นชุดฝึกและทดสอบ
X_train, X_test, y_train, y_test = train_test_split(X, y_class, test_size=0.5, random_state=0)

# เทรนและทำนายผลด้วยแต่ละโมเดล
for name, clf in zip(names, classifiers):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    score = accuracy_score(y_test, y_pred)  # ใช้ accuracy score สำหรับการจำแนกประเภท
    print(f"{name}: {score:.4f}")
```
